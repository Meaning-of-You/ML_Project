### 데이터 불러오기

```python
# pymysql import 
import pymysql
```

```python
# mysql 연결
conn = pymysql.connect(host = '', # DB IP (local에 접속하는 경우 : localhost)
                     port = 3306, # 포트 번호
                     user = '',
                     passwd = '', # mySQL 접속 계정 패스워드
                     db = 'movie', # 데이터베이스명
                     charset='utf8') # 한글 깨짐 방지
```

```python
# DB 조회 결과를 dictionary로 저장
curs = conn.cursor(pymysql.cursors.DictCursor)
```

```python
# movie 데이터 조회
sql = '''
SELECT * FROM movie
''' 
curs.execute(sql)
```

```python
# 데이터 조회 결과를 result에 저장
result = curs.fetchall()
    
print(result[0])
```

```python
# pandas 라이브러리 import
import pandas as pd
```

```python
# 데이터 프레임으로 result 결과 저장
df = pd.DataFrame(result)
print(df.columns) # columns
print(df.head())
```

```python
# 상위 3개의 행만 조회
df.head(3)
```

```python
# 데이터프레임 역순으로 뒤집기
df.loc[:,::-1].head(3)
```

```python
# 역순으로 뒤집어서 새로 저장
df = df.loc[:,::-1]
```
